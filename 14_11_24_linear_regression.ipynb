```python
#implement linear regressionusing single variable for a sample stored as csv
import pandas as pd 
from sklearn import linear_model
import matplotlib.pyplot as plt

df=pd.read_csv('homeprices.csv')
```

```python
df
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='blue',marker='*')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[6800]])
```

```python
#implement linear regression using multiple variable for a sample stored as a homeprices.csv
import pandas as pd 
from sklearn import linear_model
import matplotlib.pyplot as plt

df=pd.read_csv('homeprices_multiple.csv')
df
```

```python
df.bedrooms.median()
```

```python
df.bedrooms=df.bedrooms.fillna(df.bedrooms.median())
```

```python
x=df.iloc[:,[0,1,2]].values
y=df.iloc[:,3].values
reg=linear_model.LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2600,3,30]])
```
